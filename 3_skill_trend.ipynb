```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
import seaborn as sns
from adjustText import adjust_text
from matplotlib.ticker import PercentFormatter
```

```python
# load data
df = pd.read_csv('data/jobData.csv.gz')

# clean data
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda val: ast.literal_eval(val) if pd.notna(val) else val)
df
```

```python
df_DE_ID = df[(df['job_title_short'] == 'Data Engineer') & (df['job_country'] == 'Indonesia')].copy()
df_DE_ID['job_posted_month_no'] = df_DE_ID['job_posted_date'].dt.month

df_DE_ID
```

```python
df_explode = df_DE_ID.explode('job_skills')

df_explode
```

```python
df_pivot = df_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)
df_pivot
```

```python
pivot_copy = df_pivot.copy(deep=False)
pivot_copy.loc['Total'] = df_pivot.sum()
top_5_skills = pivot_copy.loc['Total'].sort_values(ascending=False).head(5).index
top_5_skills  
```

```python
# for percent calculation
df_total_permonth = df_DE_ID.groupby('job_posted_month_no').size()
df_total_permonth
```

```python
df_percent = df_pivot.div(df_total_permonth/100, axis=0)
df_percent
```

```python
df_percent = df_percent.reset_index()
df_percent['job_posted_month'] = df_percent['job_posted_month_no'].apply(lambda val: pd.to_datetime(val, format='%m').strftime('%b'))
df_percent = df_percent.set_index('job_posted_month')
df_percent = df_percent.drop(columns='job_posted_month_no')
df_percent
```

```python
df_plot = df_percent[top_5_skills]
sns.set_theme(style='ticks')
sns.lineplot(data=df_plot, dashes=False, palette='tab10')
sns.despine()

plt.title('Trending Top Skills for Data Engineer in Indonesia')
plt.xlabel('2025')
plt.ylabel('skills trend in Job Postings (%)')
plt.legend().remove()

# we can do this because we already sort descending
# with this code the text will have minimum of 2.5 distance
sorted_dec = df_plot.iloc[-1].sort_values(ascending=False)
minimum_dist = 2.5
prev_val = -99
texts = []
for idx in range(5):
    cur_val = sorted_dec.iloc[idx] 
    if not prev_val == -99:
        diff = prev_val - cur_val
        cur_val = cur_val - (minimum_dist - diff )if diff < minimum_dist else cur_val
    texts.append(plt.text(11,cur_val , df_plot.columns[idx]))
    prev_val = cur_val 
adjust_text(texts, arrowstyle='->', color='gray', lw=0.5)

ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))

plt.show()
```
